```python
corr_mat = pd.DataFrame(df.corr())
sns.set(font_scale = 1.2)
fig, ax = plt.subplots(figsize=(15,20)) 
heat_map = sns.heatmap(corr_mat, 
                       annot=True, 
                      yticklabels=a,
                      xticklabels=a)
plt.show()
```

```python
corr_mat = np.corrcoef(df.values.T)
sns.set(font_scale = 1.2)
fig, ax = plt.subplots(figsize=(15,20)) 
heat_map = sns.heatmap(corr_mat, 
                       annot=True, 
                      yticklabels=a,
                      xticklabels=a)
plt.show()
```
